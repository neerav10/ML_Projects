```python
dataset = pd.read_csv('dataset_Lregression_w_multiple_variables.csv')
print(dataset.shape)
print(dataset.head(5))
dataset.describe()
```

# Finding the missing values

```python
dataset.columns[dataset.isna().any()]
```

# Replacing missing values with mean

```python
dataset.hours = dataset.hours.fillna(dataset.hours.mean())
```

# Segregating Dataset into X and Y

# New Section

```python
X = dataset.iloc[:, :-1]
print(X.shape)
X
```

```python
Y = dataset.iloc[:, -1]
print(Y.shape)
Y
```

Training

```python
model = LinearRegression()
model.fit(X, Y)
```
